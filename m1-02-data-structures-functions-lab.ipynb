```python
import random
import m1_02_data_structures_functions as sf
```

```python
def generate_raw_data(num_entries=200):
    random.seed(42) 
    categories = ["Technical", "Billing", "Hardware", "Account Access"]
    category_weights = [0.5, 0.2, 0.2, 0.1] 
    data = []
    for i in range(num_entries):
        ticket = {
            "ticket_id": 1000 + i,
            "customer_id": random.randint(1, 50),
            "category": random.choice(random.choices(categories, weights=category_weights, k=10)),
            "resolution_minutes": random.choice([None, "unknown", 15, 30, 45, 120, 300]),
            "is_escalated": random.choice([True, False])
        }
        if random.random() > 0.8:
            ticket["category"] = ticket["category"].upper() + "  "
            
        data.append(ticket)
    return data

raw_logs = generate_raw_data(200)
```

```python
print(f"Total records: {len(raw_logs)}")
print("First 5 entries:", raw_logs[:5])
```

```python
#We used random.seed here to make sure the data is the same for everyone. We intentionally put None and string values like "unknown" into the resolution_minutes field to simulate a real-life “dirty data” scenario.
```
